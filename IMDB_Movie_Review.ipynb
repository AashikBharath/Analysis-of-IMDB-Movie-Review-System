```python
## AASHIK BHARATH EGNYA VARAHAN
## R00182866

## IMDB MOVIE REVIEW APPLICATION
# Importing the necessary packages

# Pandas and Numpy packages helps to import the dataset and
# analyse the dataset on various operations
import pandas as pd
import numpy as np
# matplotlib package helps to plot the output
import matplotlib.pyplot as plot
# seaborn package provides high-level interface for drawing attractive and
# informative statistical graphics
import seaborn as sb

# Importing the dataset into jupyter notebook
imdb_movie = pd.read_csv("movie_metadata.csv")
# Removes the NA values from the whole dataset
imdb_movie.dropna()
# Formats the 'movie_title' columns by removing the last extra character present
imdb_movie['movie_title']=imdb_movie['movie_title'].map(lambda x: str(x)[:-1])

# main() function of IMDB Movie Review Application
def main():
    # Prompts user to select any one of the option
    user_input = int(input("Please select one of the following options:\n\n    1. Most successful directors or actors\n    2. Film comparison\n    3. Analyse the distribution of gross earnings\n    4. Genre Analysis\n    5. Earnings and IMDB scores\n    6. Exit\n"))
    if (user_input == 1):
        # Executes the function task1()
        task1()
    elif (user_input == 2):
        # Executes the function task2()
        task2()
    elif (user_input == 3):
        # Executes the function task3()
        task3()
    elif (user_input == 4):
        # Executes the function task4()
        task4()
    elif (user_input == 5):
        # Executes the function task5()
        task5()
    elif (user_input == 6):
        print("Thanks for using IMDB Review System!")
    else:
        print("Invalid Input! Try Again!")
        main()
    
def task1():
    # Prompts the user to select either of two options
    user_input_top = int(input(" 1. Top Directors\n 2. Top Actors\n"))
    if(user_input_top == 1):
        # Exectues topDirectors() function
        topDirectors()
    elif(user_input_top == 2):
        # Exectues topActors() function
        topActors()
    else:
        print("Invalid Input! Try Again!")
        # Again calls the task1() function since invalid input is given
        task1()

def task2():
    # Asks the user for two film names to compare
    user_film1_input = str(input("Enter the first film name: "))
    user_film2_input = str(input("Enter the second film name: "))
    # Checks whether the given inputs are valid
    is_Film_Invalid = isFilmValid(user_film1_input, user_film2_input, imdb_movie)
    if is_Film_Invalid==True:
        # Executes task2_sub() function if the given inputs are valid
        task2_sub(is_Film_Invalid, user_film1_input, user_film2_input, imdb_movie)
    else:
        print("Invalid Inputs! Try Again!")
        # If the inputs are not valid, task2() is again called
        task2()
    
def task2_sub(is_Film_Invalid, user_film1_input, user_film2_input, imdb_movie):
    # Prompts user for an input to display certain type of comparison between two films
    filmComparison = int(input("1. IMDB Scores \n2. Gross Earnings \n3. Movie Facebook Likes \n4. Main Menu \n"))
    if filmComparison==1 or filmComparison==2 or filmComparison==3 or filmComparison==4:
        # Compares and plots a graph depending on the option selected by the user
        compareFilms(filmComparison, user_film1_input, user_film2_input, imdb_movie, is_Film_Invalid)
    else:
        print("Invalid Inputs! Try Again!")
        # task2_sub() function is called if the inputs are not valid
        task2_sub()
    
def task3():
    # User is prompted to enter two valid year inputs
    user_input_startyear = int(input("Enter the start year: "))
    user_input_endyear = int(input("Enter the end year: "))
    # Checks whether the given inputs are valid
    is_Year_Invalid = isYearInvalid(user_input_startyear, user_input_endyear, imdb_movie)
        
    if is_Year_Invalid==True:
        # Executes the below function if the given inputs are valid
        grossEarningDistributions(imdb_movie, user_input_startyear, user_input_endyear)
    else:
        print("Invalid Year Inputs! Try Again!")
        # task3() function is called if the inputs are invalid
        task3()
            
def task4():
    # Executes the below function
    genreAnalysis()
    
def task5():
    # Executes the below function
    earningsAndIMDBScore()
 
# Function to display Top Directors
def topDirectors():
    # User is prompted to enter the number of Top Directors to be displayed
    topDirectors_input = int(input("Enter the no. of Top Directors: "))
    # Performs error checking to confirm the input given is as per requirements
    if(topDirectors_input>0 and topDirectors_input<=len(imdb_movie.director_name.unique())):
        # Groups the data by 'Director name' and finds the total gross film earnings for that director
        topDirectors_grouped = imdb_movie.groupby('director_name', as_index=False)['gross'].sum()
        # Sorts the grouped data in descending order based on total gross earnings
        topDirectors_sorted = topDirectors_grouped.sort_values(by='gross',ascending=[False])
        # Displays a list of Top Directors as mentioned by the user
        topDirectors_head = topDirectors_sorted.head(topDirectors_input)
        # Plots a bar graph on the subset data
        dir_plot=sb.barplot(x="gross", y="director_name", data=topDirectors_head)
        dir_plot.set(xlabel="Gross in Billions")
        dir_plot.set(ylabel="Director Name")
        # Function of matplotlib package to display the graph
        plot.show()
        # The user is again taken back to the main menu once the results are produced
        main()   
    else:
        print("Invalid Input! Please try again!")
        # The below function is caled again if the inputs are not valid
        topDirectors()

# Function to display Top Actors
# Performs the same operation as topDirectors() function
def topActors():
    topActors_input = int(input("Enter the no. of Top Actors: "))
    count_top_Actors=imdb_movie.groupby(['actor_1_name','actor_2_name','actor_3_name'])['gross'].sum()
    topActors_count = count_top_Actors.count()
    if(topActors_input>0 and topActors_input<=topActors_count):
        topActors_groupby = imdb_movie.groupby(['actor_1_name','actor_2_name','actor_3_name'], as_index=False)['gross'].sum()
        topActors_sorted = topActors_groupby.sort_values(by='gross',ascending=[False])
        topActors_head = topActors_sorted.head(topActors_input+1)
        actor_plot=sb.barplot(x="gross", y="actor_1_name", data=topActors_head)
        actor_plot.set(xlabel="Gross in Billions")
        actor_plot.set(ylabel="Actor Name")
        plot.show()
        main()
    else:
        print("Invalid Input! Please try again!")
        topActors()

# Function to check whether the given inputs are valid or not
def isFilmValid(user_film1_input, user_film2_input, imdb_movie):
    # Checks whether the user given inputs of films are available in dataset or not
    if user_film1_input.lower() in imdb_movie['movie_title'].str.lower().unique() and user_film2_input.lower() in imdb_movie['movie_title'].str.lower().unique():
        return True
    else:
        return False

# Compares the two films given as inputs 
def compareFilms(filmComparison, user_film1_input, user_film2_input, imdb_movie, is_Film_Invalid):
    compare_films=None
    # Subsets the data limited to two film names given as input
    compare_films=imdb_movie[(imdb_movie.movie_title.str.lower()==user_film1_input.lower())|(imdb_movie.movie_title.str.lower()==user_film2_input.lower())]
    
    # Plots a bar graph for two films against their respective IMDB scores
    if(filmComparison==1):
        imdb_plot=sb.barplot(x="movie_title",y="imdb_score", data=compare_films)
        imdb_plot.set(xlabel="Movie Title")
        imdb_plot.set(ylabel="IMDB Score")
        plot.show()
        # Takes back the user to sub menu and ask the user for new comparison with the same two films
        task2_sub(is_Film_Invalid, user_film1_input, user_film2_input, imdb_movie)
    
    # Plots a bar graph for two films against their respective total gross film earnings
    elif(filmComparison==2):
        gross_earnings_plot=sb.barplot(x="movie_title",y="gross", data=compare_films)
        gross_earnings_plot.set(xlabel="Movie Title")
        gross_earnings_plot.set(ylabel="Gross Earnings")
        plot.show()
        task2_sub(is_Film_Invalid, user_film1_input, user_film2_input, imdb_movie)
    
    # Plots a bar graph for two films against their respective total facebook likes
    elif(filmComparison==3):
        movie_fb_plot=sb.barplot(x="movie_title",y="movie_facebook_likes", data=compare_films)
        movie_fb_plot.set(xlabel="Movie Title")
        movie_fb_plot.set(ylabel="Movie FB Likes")
        plot.show()
        task2_sub(is_Film_Invalid, user_film1_input, user_film2_input, imdb_movie)
        
    else:
        # Takes the user to main menu if none of the above option is selected
        main()

# Check whether the given input years are valid or not
def isYearInvalid(user_input_startyear, user_input_endyear, imdb_movie):
    # Checks if the startyear is less than the endyear
    if user_input_startyear < user_input_endyear:
        # Checks if the startyear and endyear are in IMDB Movie dataset and within the range
        if user_input_startyear in imdb_movie['title_year'].unique() and user_input_endyear in imdb_movie['title_year'].unique():
            return True
        else:
            return False
    else:
        return False

# Plots line graph for a range of title years against its max, min and avg values
def grossEarningDistributions(imdb_movie, user_input_startyear, user_input_endyear):
    # Subsets the data based on the requirement
    yearData = imdb_movie[(imdb_movie['title_year']>=user_input_startyear) & (imdb_movie['title_year']<=user_input_endyear)]
    # Groups the data by film title year and finds the aggregate function 'min','mean' and 'max' for gross film earnings
    yearGross = yearData.groupby('title_year').agg({'gross':['min','mean','max']})
    yearGross.columns = yearGross.columns.droplevel(level=0)
    # Plots line graph for the different aggregate functions
    gross_earnings_plot=yearGross.plot.line()
    gross_earnings_plot.set(xlabel="Title Year")
    gross_earnings_plot.set(ylabel="Gross Earnings")
    plot.show()
    # Display main menu options
    main()
  
# Display a list of available Genres
def genreAnalysis():
    # List of available genres
    genre_list = imdb_movie['genres'].str.split("|")
    unique_list = [genre for genres in genre_list for genre in genres]
    unique_list = np.unique(unique_list)
    print("List of Genres Analysis: \n")
    for genre in unique_list:
        print(genre)
    # Executes the below function after displaying the genre list
    meanIMDBScore()

# Calculates the average IMDB Score for the specific genre
def meanIMDBScore():
    # This function calculates the mean IMDB Score
    # User is given an option to enter the Genre for which average IMDB score needs to be calculated
    user_choice_genre = input("Enter a Genre for IMDB Score: ")
    user_choice_genre = user_choice_genre.lower()
    genre_list = imdb_movie['genres'].str.lower().str.split("|")
    unique_list = [genre for genres in genre_list for genre in genres]
    unique_list = np.unique(unique_list)
    
    # Checks if the given input is available in genre list
    if user_choice_genre in unique_list:
        # Calculates the mean IMDB score for the specific genre
        genIndex = imdb_movie['genres'].str.lower().str.contains(user_choice_genre.lower())
        genData = imdb_movie[genIndex]
        score = round(genData['imdb_score'].mean(),2)
        print("IMDB score for",user_choice_genre,"is:",score) 
        main()
    else: 
        print("Invalid Genre! Try Again with different input!")
        meanIMDBScore()

# Finds the relationship between various numerical variables to build a model
def earningsAndIMDBScore():
    # Finds relation between numerical variables
    # Sorts the numeric variables
    numTypes = ['int64','float64']
    numData = imdb_movie.select_dtypes(include=numTypes)
    # Finds the correlation relation
    corr=numData.corr()
    print("Correlation btween imdb_score and other numerical variables: \n")
    print(corr['imdb_score'])
    # Plots a correlation relation between various numerical variables using seaborn package
    sb.heatmap(corr)
    plot.show()
    
    # Scatterplot is generated for every numerical variable against IMDB Score variable for model building & prediction
    for i, col in enumerate(numData.columns):
        if col!='imdb_score':
            plot.figure(i)
            sb.scatterplot(x=col, y='imdb_score', data=numData, color='Red')
            plot.show()
    # Takes the user back to main menu option
    main()

    
# Executes the application
main()
```
